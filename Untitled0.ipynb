```python
import pandas as pd
import numpy as np
import catboost
from catboost import CatBoostRegressor
from catboost import Pool, CatBoostClassifier
from sklearn.model_selection import GridSearchCV
import re
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, normalize
from catboost.text_processing import Dictionary, Tokenizer
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.decomposition import PCA
from sklearn import decomposition # подключим модуль
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
train=pd.read_csv("/content/drive/MyDrive/Data_science/НЕЙРОСЕТИ/Catboost/мошеннические операции/train (3).csv",sep=";")
```

```python
test=pd.read_csv("/content/drive/MyDrive/Data_science/НЕЙРОСЕТИ/Catboost/мошеннические операции/test (3).csv",sep=';')
```

```python
sub=pd.read_csv("/content/drive/MyDrive/Data_science/НЕЙРОСЕТИ/Catboost/мошеннические операции/sample_submission (3).csv",sep=';',index_col=False)
```

```python
train["STATE"] = train["STATE"].apply(lambda item: item.lower())
train["DOMAIN"] = train["DOMAIN"].apply(lambda item: item.lower())
```

```python
test["STATE"] = test["STATE"].apply(lambda item: item.lower())
test["DOMAIN"] = test["DOMAIN"].apply(lambda item: item.lower())
```

```python
train=pd.get_dummies(train, columns=['TRN_TYPE'], drop_first= True)
```

```python
le = preprocessing.LabelEncoder()
X_domain_train = le.fit_transform(train["DOMAIN"])
X_state_train = le.fit_transform(train["STATE"])
```

```python
le = preprocessing.LabelEncoder()
X_domain_test = le.fit_transform(test["DOMAIN"])
X_state_test = le.fit_transform(test["STATE"])
```

```python
train["DOMAIN"]=X_domain_train
train["STATE"]=X_state_train
```

```python
test["DOMAIN"]=X_domain_test
test["STATE"]=X_state_test
```

```python
test.drop(["id"],axis=1,inplace=True)
```

```python
target = "TRN_TYPE_LEGIT"
y = train.loc[:, target]
X = train.drop(columns=[target])
```

```python
test.drop(["DOMAIN","STATE"],axis=1,inplace=True)
```

```python
X.drop(["DOMAIN","STATE"],axis=1,inplace=True)
```

```python
X_plc=X.dropna()
```

```python
from sklearn.manifold import TSNE # здесь подключаем метод TSNE
tsne = TSNE(random_state=17) # создаем модель TSNE, по умолчанию 2 компоненты

X_tsne = tsne.fit_transform(X) # обучаем ее и строим проекцию данных на компоненты
```

```python
pca = decomposition.PCA(n_components=2)
X_centered = X_plc - X_plc.mean(axis=0) # отцентруем данные
pca.fit(X_centered) # обучим PCA
X_pca = pca.transform(X_centered)
```

```python
pca = decomposition.PCA().fit(X)

plt.figure(figsize=(10,7))
# будем рисовать процент "объясненной" дисперсии добавляя новые компоненты
evr=np.cumsum(pca.explained_variance_ratio_)
plt.plot(evr, color='k', lw=2, marker='o')
plt.xlabel('Number of components')
plt.ylabel('Total explained variance')
plt.xlim(0, 63)
plt.yticks(np.arange(0, 1.1, 0.1))

cmp=np.where(evr>0.9)[0][0] # здесь набeрется 90%
print('90% variance at component № ', cmp+1)

plt.axvline(cmp, c='b')
plt.axhline(0.9, c='r')
plt.show();
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_plc, y, test_size=.1,
                                                    stratify=y,
                                                    random_state=42)
```

```python
X_pca
```

```python
cat_features = ['VAR3', 'TRN_AMT', 'TOTAL_TRN_AMT']
```

```python
text_features = ['DOMAIN',"STATE"]
```

```python
train_pool = Pool(
    x_train,
    y_train,
    #text_features=text_features,
    #cat_features=cat_features,
)


valid_pool = Pool(
    x_test,
    y_test,
    #text_features=text_features,
    #cat_features=cat_features,
)
```

```python
cat = CatBoostClassifier(
    #cat_features=cat_features,
    #text_features=text_features,
    iterations=1500,
    l2_leaf_reg=0.12,
    depth=6,
    learning_rate=0.8,
    eval_metric="TotalF1",
    use_best_model=True)
#ExtraTreesRegressor
```

```python
grid = {'learning_rate': [0.03, 0.1],
        'depth': [4, 6, 10],
        'l2_leaf_reg': [1, 3, 5, 7, 9]}

grid_search_result = cat.grid_search(grid,
                                       X=x_train,
                                       y=y_train,
                                       plot=True)
```

```python
cat.fit(train_pool, eval_set=valid_pool, use_best_model=True)
```

```python
predict = cat.predict(test)
```
